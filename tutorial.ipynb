# Example code for loading saved data

```python
import numpy as np

data = np.load('./embeddings.npz', allow_pickle=True)

# To see the array names stored in the npz file:
print(data.files)

# For example, to access the embeddings and ids arrays:
embeddings = data['embeddings']
ids = data['ids']

print("embeddings shape:", embeddings.shape)
print("ids shape:", ids.shape)
print("First sequence ID:", ids[0])
print("First sequence embedding shape:", embeddings[0].shape if embeddings.ndim==3 else embeddings[0].shape)
```
